```python
import sqlite3
import pandas as pd
```

```python
#Initiate database and cursor
path = r"C:\Users\cailopes\Downloads\AmazonBooks_2018.db"
db = sqlite3.connect(path)
cur = db.cursor()
```

```python
#JUST RUN ONE TIME
#Deleting NULLS and empty books

delete = """
    DELETE FROM Books
    
    WHERE Book_Content = '' OR Book_Content IS NULL
"""

cur.execute(delete)
db.commit()
```

```python
#Define the index of the book (THIS IS IMPORTANT!!)
index = 204
```

```python
#Select book from that index
query = """
    SELECT Book_Content
    
    FROM Books
    
    WHERE Id = {};

""".format(index)

cur.execute(query)
book = cur.fetchall()[0][0].split("<p>")

#Print book and search for the string where the Book actually starts
for piece in book:
    print(piece)
```

```python
#Put the string you copied here and check if the book is correct now

my_string = r"""The patrol blimp Midnight Dancer swam through"""

for i in range(0, len(book)):
    if my_string in book[i]:
        start = i
        break

kindle_string = r'End of this sample Kindle book.Enjoyed the preview?Buy with 1-ClickorSee details for this book in the Kindle Store'
end = len(book)-1

for j in range(0, len(book)):
    if kindle_string in book[j]:
        end = j
        break
        
fixed_book = book[start:end]

for piece in fixed_book:
    print(piece)
```

```python
#turn Book into one single string again
s_book = ""
for piece in fixed_book:
    s_book = s_book + piece + "<p>"
    
#update book!

update = """
    UPDATE Books
    
    SET Book_Content = ?
    
    WHERE Id = {};
""".format(index)

cur.execute(update, [s_book])
db.commit()
```
